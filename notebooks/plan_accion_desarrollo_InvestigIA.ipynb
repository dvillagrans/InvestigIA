# 1. Planteamiento del problema

La formulación de un proyecto de investigación de posgrado puede modelarse como un proceso iterativo de reducción de incertidumbre. En las etapas iniciales, el estudiante dispone de un conjunto amplio de intereses, conocimientos parciales y restricciones institucionales. El objetivo del proceso consiste en transformar dicho estado inicial en una representación estructurada del proyecto:

$$
\mathcal{P} =
\left\{
q,\;
o_g,\;
O_e,\;
h,\;
m,\;
j,\;
a,\;
r
\right\}
$$

donde:

- $q$: pregunta de investigación,
- $o_g$: objetivo general,
- $O_e$: conjunto de objetivos específicos,
- $h$: hipótesis o supuesto de trabajo,
- $m$: estrategia metodológica,
- $j$: justificación,
- $a$: alcance,
- $r$: resultados esperados.

En la práctica, esta transformación presenta tres dificultades principales:

1. **Alta entropía temática:** el estudiante inicia con múltiples líneas posibles y baja claridad respecto al alcance.
2. **Costo cognitivo de la revisión de literatura:** la búsqueda y lectura inicial no se encuentra integrada en un flujo metodológico.
3. **Dependencia de retroalimentación experta:** la convergencia hacia un protocolo viable depende de ciclos prolongados con el asesor.

InvestigIA se propone como un sistema de apoyo que reduzca dicha incertidumbre mediante agentes especializados, recuperación aumentada por generación y validación estructurada.

# 2. Preguntas de investigación

La prueba de concepto se articula alrededor de las siguientes preguntas:

### PI-1

**¿En qué medida una arquitectura multiagente local puede transformar información académica no estructurada del estudiante en un protocolo preliminar con secciones metodológicamente coherentes?**

### PI-2

**¿Qué nivel de trazabilidad bibliográfica puede alcanzarse mediante un esquema RAG local sin utilizar servicios externos de inferencia o recuperación?**

### PI-3

**¿Es posible ejecutar un flujo completo de perfilado, delimitación, recuperación documental y generación en hardware institucional con latencias compatibles con una interacción conversacional?**

### PI-4

**¿Qué combinación de validaciones deterministas y revisión basada en LLM permite reducir inconsistencias internas en el protocolo generado?**

### PI-5

**¿Qué componentes del flujo aportan mayor valor percibido por el estudiante y cuáles requieren intervención obligatoria del asesor?**

# 3. Hipótesis de trabajo

### H1 — Factibilidad funcional

Una arquitectura compuesta por agentes especializados con contratos de entrada y salida estructurados puede producir un protocolo preliminar completo a partir de un perfil académico y un corpus local.

### H2 — Reducción de retrabajo

El uso de preguntas adaptativas y recuperación contextual reducirá el número de iteraciones necesarias para obtener una primera versión coherente del proyecto.

### H3 — Trazabilidad

La separación explícita entre contenido recuperado y contenido generado permitirá mantener referencias verificables y disminuir la aparición de citas inexistentes.

### H4 — Operación local

Un modelo cuantizado de la familia Gemma 4, servido mediante LM Studio o un motor compatible, podrá ejecutar el flujo de la PoC dentro de límites razonables de memoria y tiempo de respuesta.

Estas hipótesis no se consideran validadas por el simple funcionamiento del prototipo; deberán contrastarse mediante métricas técnicas y evaluación con usuarios.

# 5. Modelo conceptual del sistema

El sistema puede representarse como una composición de operadores:

$$
\mathcal{F}
=
V \circ G \circ R \circ D \circ M \circ P
$$

donde:

- $P$: perfilado del estudiante,
- $M$: matching de temas y asesores,
- $D$: delimitación metodológica,
- $R$: recuperación de evidencia,
- $G$: generación del protocolo,
- $V$: validación.

Para una entrada $x$, el sistema genera:

$$
y = \mathcal{F}(x)
$$

La utilidad de la salida puede expresarse como:

$$
U(y) =
\alpha C +
\beta T +
\gamma K +
\delta E
$$

donde:

- $C$: completitud,
- $T$: trazabilidad,
- $K$: coherencia interna,
- $E$: eficiencia computacional,
- $\alpha,\beta,\gamma,\delta$: pesos ajustables conforme al criterio institucional.

En esta fase no se pretende optimizar globalmente $U(y)$, sino verificar que cada componente alcance un umbral mínimo aceptable.

# 8. Formalización del módulo de matching

El matching inicial se implementará mediante una función de puntuación:

$$
S(a_i,s)
=
w_1 I_i +
w_2 E_i +
w_3 H_i +
w_4 D_i
$$

donde:

- $a_i$: asesor o línea candidata,
- $s$: perfil del estudiante,
- $I_i$: similitud de intereses,
- $E_i$: afinidad con la experiencia previa,
- $H_i$: compatibilidad de habilidades,
- $D_i$: disponibilidad o carga,
- $w_j$: pesos configurables.

Para conservar interpretabilidad, los pesos deberán satisfacer:

$$
w_j \geq 0,
\qquad
\sum_{j=1}^{4} w_j = 1
$$

La PoC podrá utilizar similitud léxica o semántica simple. Todavía no se plantea un recomendador entrenado; el objetivo es producir un ranking interpretable, no una clasificación opaca.

# 9. Formalización del módulo RAG

Para una consulta $q$, cada fragmento $d_i$ se representa mediante un vector de embedding $e(d_i)$. La recuperación inicial se realizará mediante similitud coseno:

$$
\operatorname{sim}(q,d_i)
=
\frac{
e(q)\cdot e(d_i)
}{
\lVert e(q)\rVert_2
\lVert e(d_i)\rVert_2
}
$$

Se recuperarán los $k$ fragmentos con mayor puntuación:

$$
D_k(q)
=
\operatorname{TopK}_{d_i}
\left(
\operatorname{sim}(q,d_i)
\right)
$$

La respuesta generada deberá estar condicionada por $D_k(q)$ y cada fragmento conservará:

- identificador,
- título,
- autores,
- año,
- página,
- puntuación,
- texto recuperado.

Cuando la evidencia recuperada sea insuficiente, la respuesta deberá marcarse explícitamente como incompleta.

# 10. Diseño de agentes

| Agente | Función | Entrada | Salida |
|---|---|---|---|
| Perfil | Estructura datos del estudiante | Texto y formulario | Perfil JSON |
| Matching | Ordena líneas y asesores | Perfil y catálogo | Ranking explicado |
| Metodológico | Delimita problema y objetivos | Tema seleccionado | Esquema de investigación |
| Investigador | Recupera evidencia | Consulta estructurada | Fragmentos y referencias |
| Redacción | Integra secciones | Esquema + evidencia | Protocolo |
| Revisión | Evalúa consistencia | Protocolo | Observaciones y puntajes |

Cada agente deberá cumplir:

- contrato de datos,
- prompt versionado,
- máximo de iteraciones,
- manejo de errores,
- logging,
- prueba unitaria o de integración.

# 11. Métricas de evaluación

## 11.1 Completitud

$$
C =
\frac{
N_{\mathrm{secciones\ completas}}
}{
N_{\mathrm{secciones\ requeridas}}
}
$$

## 11.2 Trazabilidad

$$
T =
\frac{
N_{\mathrm{afirmaciones\ con\ fuente}}
}{
N_{\mathrm{afirmaciones\ que\ requieren\ fuente}}
}
$$

## 11.3 Coherencia estructural

Se evaluará mediante reglas como:

- correspondencia entre pregunta y objetivo general,
- correspondencia entre objetivos específicos y metodología,
- presencia de variables o unidades de análisis,
- consistencia temporal del cronograma,
- compatibilidad entre hipótesis y tipo de estudio.

## 11.4 Rendimiento

La latencia total se modelará como:

$$
L_{\mathrm{total}}
=
L_P +
L_M +
L_D +
L_R +
L_G +
L_V
$$

donde $L_i$ representa la latencia observada en cada módulo.

También se registrará:

- RAM máxima,
- VRAM máxima,
- tokens por segundo,
- tiempo de indexación,
- tiempo de recuperación,
- tamaño del corpus.

# 18. Amenazas a la validez

## Validez interna

- El corpus controlado puede favorecer artificialmente la recuperación.
- Los prompts pueden estar sobreajustados a los casos de prueba.
- La evaluación inicial puede depender de criterios subjetivos.

## Validez externa

- Los resultados con estudiantes del laboratorio no garantizan generalización a otras disciplinas.
- El desempeño del modelo puede variar significativamente según cuantización y hardware.
- Una PoC de diez días no representa el comportamiento de una plataforma multiusuario.

## Validez de constructo

- “Calidad del protocolo” no debe reducirse únicamente a completitud textual.
- La coherencia automática no sustituye la evaluación metodológica de un experto.
